### 1. Hvor mange mennesker har fødselsdag d. 07-01 i 2020

```python
import pandas as pd 
import numpy as np

url = 'https://api.statbank.dk/v1/data/FOLK3/CSV?delimiter=Semicolon&FDAG=D07&FMAANED=001&FODAAR=1992&Tid=*'
df = pd.read_csv(url,sep=';')
data = np.array(df)

print('Antal mennesker der har fødselsdag samme dag som mig selv er: {}'.format(data[12][4]))
```

### 2. Hvad er den procentvise stigning af mennesker med samme fødselsdag som mig selv fra 2008 til 2020?

```python
d1 = data[(data[:,3] == 2008)][0][4]
d2 = data[(data[:,3] == 2020)][0][4]
print('Den procentvise stigning af mennesker med samme fødselsdag som mig selv er: {}%'.format((d2-d1) / d1 * 100))
```

### 3. Vis et bar plot over udvikling af din fødselsdag fra 2008 til 2020

```python
import matplotlib.pyplot as plt

birthdayCount = {}
for row in data:
    birthdayCount[row[3]] = row[4]

plt.figure(figsize=(20, 5))
plt.bar(birthdayCount.keys(), birthdayCount.values(), align='center', alpha=0.5,)
plt.ylabel('Antal mennesker')
plt.xlabel('Årstal')
plt.title('Mennesker med fødselsdag d. 07-01 i tiden 2008 til 2020')
```
