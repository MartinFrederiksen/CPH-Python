```python
import Module.Module as module
import matplotlib.pyplot as plt

texts = ['https://www.gutenberg.org/files/1342/1342-0.txt', 'https://www.gutenberg.org/files/84/84-0.txt', 'http://www.gutenberg.org/cache/epub/25525/pg25525.txt', 'http://www.gutenberg.org/cache/epub/1635/pg1635.txt', 'https://www.gutenberg.org/files/1080/1080-0.txt', 'http://www.gutenberg.org/cache/epub/2542/pg2542.txt', 'https://www.gutenberg.org/files/11/11-0.txt', 'http://www.gutenberg.org/cache/epub/514/pg514.txt', 'https://www.gutenberg.org/files/215/215-0.txt', 'http://www.gutenberg.org/cache/epub/844/pg844.txt']
m = module.Module(texts)

#m.download('https://www.gutenberg.org/files/98/98-0.txt', 'testText')
#m.multiDownload()
#print([u for u in m])
#print([u for u in m.fileListGenerator()])

data = {}
for url in texts:
    with open(url.split('/')[-1], 'r') as file:
        data[url.split('/')[-1]] = m.avgVowels(file.read())

#print(m.hardestRead())

dataSorted = {k: v for k, v in sorted(data.items(), key=lambda item: item[1], reverse=True)}
plt.figure(figsize=(20, 5))
plt.bar(dataSorted.keys(), dataSorted.values(), align='center', alpha=0.5,)
plt.ylabel('Avg Vowels')
plt.xlabel('Txt name')
plt.title('Avg vowels in descending order')
```
